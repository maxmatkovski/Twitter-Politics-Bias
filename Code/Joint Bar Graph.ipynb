```python
import numpy as np 
import matplotlib.pyplot as plt

values = [1,2,3,4,5]
republicans = [547, 90, 134, 75, 155]
democrats = [572, 82, 129, 75, 143]

X_axis = np.arange(len(values))
  
plt.bar(X_axis - 0.2, republicans, 0.4, label = 'Republican', color='red')
plt.bar(X_axis + 0.2, democrats, 0.4, label = 'Democrat')
  
plt.xticks(X_axis, values)
plt.xlabel("Republican vs Democrat Tweet Sentiment Scores")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
